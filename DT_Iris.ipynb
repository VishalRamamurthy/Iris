```python
# 3. Split the data into train and test
from sklearn.model_selection import train_test_split
train, test = train_test_split(iris, test_size = 0.2, random_state = 1)
# train, test = train_test_split(iris, test_size = 0.2)
```

```python
print(iris.shape)
print(train.shape)
print(test.shape)
```

```python
train.head()
```

```python
test.head()
```

```python
trainx = train.iloc[:, 0:4]
trainy = train.iloc[:, 4]
testx = test.iloc[:, 0:4]
testy = test.iloc[:, 4]
```

```python
trainx.head()
```

```python
trainy.head()
```

```python
testx.head()
```

```python
testy.head()
```

```python
# Supervised-Classification
from sklearn.tree import DecisionTreeClassifier as DT
model = DT(criterion = 'entropy')
# learn a model
model.fit(trainx,trainy)
# Make predictions
predictions = model.predict(testx)
```

```python
predictions
```

```python
testy
```

```python
accuracy = np.mean(predictions == testy)
accuracy
```
